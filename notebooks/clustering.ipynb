```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from plotnine import *
import plotnine
```

```python
user_df = pd.read_csv('../output/user_df.csv',index_col=0)
```

```python
user_df = user_df.drop('user_id_mapping',axis=1)
```

```python
user_df.shape
```

```python
# normalize for each feature across users, standerlize each features and make them comparable (like same weight)
dfMatrix = user_df.to_numpy()
scaled_df = StandardScaler().fit_transform(dfMatrix)
```

```python
# Choose optimal K using Elbow method
cost = []
for cluster in range(1, 15):
    clustering = KMeans(n_clusters=cluster, init='k-means++', random_state=42)
    clustering.fit(scaled_df)
    cost.append(clustering.inertia_)
    print('Cluster initiation: {}'.format(cluster))
 

# Converting the results into a dataframe and plotting them
df_cost = pd.DataFrame({'Cluster':range(1, len(cost) + 1), 'Cost':cost})
# Data viz
plotnine.options.figure_size = (10, 6)
(
    ggplot(data = df_cost)+
    geom_line(aes(x = 'Cluster',
                  y = 'Cost'))+
    geom_point(aes(x = 'Cluster',
                   y = 'Cost'))+
    geom_label(aes(x = 'Cluster',
                   y = 'Cost',
                   label = 'Cluster'),
               size = 10,
               nudge_y = 1000) +
    labs(title = 'Optimal number of cluster with Elbow Method')+
    xlab('Number of Clusters k')+
    ylab('Cost')+
    theme_minimal()
)
```

```python
# Fit the cluster
clustering = KMeans(n_clusters=5, init='k-means++', random_state=42)
clustering.fit(scaled_df)

# Check the cost of the clusters created
clustering.inertia_
```

```python
user_df['cluster'] = clustering.labels_
```
